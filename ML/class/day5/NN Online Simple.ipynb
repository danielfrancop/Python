```python
import time
import numpy as np
```

```python
def sigmoid(x):
    return 1 / ( 1 + np.exp(-x) )

def sigmoid_der(x):
    return x * (1 - x)

```

Set random weights 

```python
w1 = 0.1
w2 = 0.2
w3 = 0.3
w4 = 0.4
w5 = 0.5
w6 = 0.6
```

Learning data

```python
xvals = [0,0,1,1]
yvals = [0,1,0,1]

expected = [0,1,1,1]
```

Process

```python
it = 0
item = 0
flag_print = -1

while True: 
    # forward propogation
    N1val = xvals[item] * w1 + yvals[item] * w3 
    N2val = xvals[item] * w2 + yvals[item] * w4 
    
    N1Out = sigmoid(N1val)
    N2Out = sigmoid(N2val)
    
    N3val = N1Out * w5 + N2Out * w6 
    
    N3Out = sigmoid(N3val)
    
    error = expected[item] - N3Out
    
    if it % 1000 == 0:
        flag_print = 0
        print("============")
        time.sleep(1)
    
    if flag_print >= 0:    
        print (N3Out)
        flag_print +=1
        if flag_print == 4:
            flag_print = -1
    
    
    # back prop
    
    
    learning = 0.5
    
    e5 = learning * error * N1Out * sigmoid_der(N3Out)
    e6 = learning * error * N2Out * sigmoid_der(N3Out)

    
    w5 += e5
    w6 += e6
    
    herr5 = error * sigmoid_der(N3Out)*w5
    herr6 = error * sigmoid_der(N3Out)*w6
    
    e1 = learning * herr5 * xvals[item] * sigmoid_der(N1Out)
    e2 = learning * herr6 * xvals[item] * sigmoid_der(N2Out)
    e3 = learning * herr5 * yvals[item] * sigmoid_der(N1Out)
    e4 = learning * herr6 * yvals[item] * sigmoid_der(N2Out)
    
    
    
    w1 += e1
    w2 += e2
    w3 += e3
    w4 += e4
    
    
    it+=1
    item += 1
    item = item % 4

```

## Adding Bias

```python
w1 = 0.1
w2 = 0.2
w3 = 0.3
w4 = 0.4
w5 = 0.5
w6 = 0.6
w33 = 0.1
w44 = 0.1
w55 = 0.1



xvals = [0,0,1,1]
yvals = [0,1,0,1]
expected = [1,1,1,0]
bias = 1
```

```python
it = 0
item = 0
flag_print = -1
count = 0

while True: 
    # forward propogation
    N3val = xvals[item] * w1 + yvals[item] * w3 + bias * w33
    N4val = xvals[item] * w2 + yvals[item] * w4 + bias * w44
    
    N3Out = sigmoid(N3val)
    N4Out = sigmoid(N4val)
    
    N5val = N3Out * w5 + N4Out * w6 + bias * w55
    
    N5Out = sigmoid(N5val)
    
    error = expected[item] - N5Out
    
    if it % 300 == 0:
        flag_print = 0
        time.sleep(1)
        count+=1
        print("===========",count)
    
    if flag_print >= 0:    
        print (N5Out)
        flag_print +=1
        if flag_print == 4:
            flag_print = -1
    
    
    # back prop - Using Stochastic Gradient descent 
    # use one observation at a time and update all weights
    
    
    learning = 0.3
    
    e5 = learning * error * N3Out * sigmoid_der(N5Out)
    e6 = learning * error * N4Out * sigmoid_der(N5Out)

    e55 = learning * error * bias * sigmoid_der(N5Out)
    
    w5 += e5
    w6 += e6
    w55 += e55
    
    herr5 = error * sigmoid_der(N5Out) * w5
    herr6 = error * sigmoid_der(N5Out) * w6
    
    e1 = learning * herr5 * xvals[item] * sigmoid_der(N3Out)
    e2 = learning * herr6 * xvals[item] * sigmoid_der(N4Out)
    e3 = learning * herr5 * yvals[item] * sigmoid_der(N3Out)
    e4 = learning * herr6 * yvals[item] * sigmoid_der(N4Out)

    
    
    e33 = learning * herr5 * bias * sigmoid_der(N3Out)
    e44 = learning * herr6 * bias * sigmoid_der(N4Out)
    
    
    w1 += e1
    w2 += e2
    w3 += e3
    w4 += e4
    
    w33 += e33
    w44 += e44
    
    it+=1
    item += 1
    if item == 4:
        item = 0

```
